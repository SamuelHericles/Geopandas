```python
import geopandas as gpd
import pandas as pd
```

```python
import os # para navegar pelos diretórios do sistema operacional
import zipfile # zipa os arquivos criados 
import shutil # apagar pasta de arquivos temporários
```

```python
# cria um diretório caso não exista.
dir = '../01.Dados/Mapas/RJ-SETOR/TEMP'
if not os.path.exists(dir):
    os.makedirs(dir)
```

```python
# Recebe os nomes dos arquivos da pasta '/DADOS'
filenames = []
for root, dirs, files in os.walk('../01.Dados/Mapas/RJ-SETOR/DADOS'):
    filenames = files
filenames
```

```python
# Coleta apenas os nomes dos arquivos .zip e armazena em uma lista.
for item in filenames:
    zip_ref = zipfile.ZipFile('../01.Dados/Mapas/RJ-SETOR/DADOS/'+item,'r')
    zip_ref.extractall(dir)
    zip_ref.close()
```

```python
filenames[0][:-4]
```

```python
# Concatena os nomes e salva em um .shp
setor = pd.concat([gpd.read_file(dir +'/'+ item[:-4] + '_setor.shp') for item in filenames],ignore_index=True)
```

```python
# Apaga os arquivos criados temporáriamente
shutil.rmtree(dir)
```

```python
setor
```

```python
type(setor)
```

```python
#Modifica o crs da base de dados
setor = setor.to_crs('+proj=utm +zone=23 +south +ellps=GRS80 +towgs84=0,0,0,0,0,0,0 +units=km +no_defs')
setor.crs
```

```python
setor.plot(color='white',edgecolor='black',figsize=(15,8))
```
