# Mapas de calor

```python
import geopandas as gpd
import folium
```

```python
rj = gpd.read_file('../01.Dados/Mapas/RJ-MINUC/RJ-MINUC.shp')
geo_dados = gpd.read_file('../01.Dados/Mapas/RJ-DATASET/DATASET.shp')
setor= gpd.read_file('../01.Dados/Mapas/RJ-SETOR/RJ-SETOR.shp')
bairro = gpd.read_file('../01.Dados/Mapas/RJ-BAIRRO/RJ-BAIRRO.shp')
```

```python
crs = {'init':'epsg:4326'}
```

```python
rj.to_crs(crs,inplace=True)
```

```python
setor.to_crs(crs,inplace=True)
```

```python
setor.fillna(0, inplace=True)
```

```python
y = rj.centroid.y.iloc[0]
x = rj.centroid.x.iloc[0]
```

```python
data = []

for i in range(len(setor)):
    data.append([setor['geometry'][i].centroid.y,
                 setor['geometry'][i].centroid.x,
                 setor['V002'][i]/setor['V002'].max()])
```

```python
data
```

```python
from folium.plugins import HeatMap
```

```python
base = folium.Map([y,x], zoom_start=11, tiles='OpenStreetMap')
HeatMap(data, name='Populao').add_to(base)
folium.LayerControl().add_to(base)
base.save('HeatMap.html')
base
```
