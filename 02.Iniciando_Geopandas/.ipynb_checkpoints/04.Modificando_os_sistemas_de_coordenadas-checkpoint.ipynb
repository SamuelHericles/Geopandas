```python
import geopandas as gpd
```

```python
rj = gpd.read_file('../01.Dados/Mapas/RJ-MINUC/RJ-MINUC.shp')
geo_dados = gpd.read_file('../01.Dados/Mapas/RJ-DATASET/DATASET.shp')
geo_dados
```

```python
rj.crs
```

```python
rj = rj.to_crs('+proj=utm +zone=23 +south +ellps=GRS80 +towgs84=0,0,0,0,0,0,0 +units=m +no_defs')
rj.crs
```

```python
geo_dados = geo_dados.to_crs('+proj=utm +zone=23 +south +ellps=GRS80 +towgs84=0,0,0,0,0,0,0 +units=m +no_defs')
geo_dados.crs
```

```python
rj.to_file('../01.Dados/Mapas/RJ-MINUC/RJ-MINUC.shp')
geo_dados.to_file('../01.Dados/Mapas/RJ-DATASET/DATASET.shp',encoding='UTF-8')
```
