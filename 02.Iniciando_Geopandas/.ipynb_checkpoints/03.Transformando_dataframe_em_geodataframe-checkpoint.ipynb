# Transformando dataframe em geodataframe

```python
import pandas as pd
import geopandas as gpd
```

```python
# read_table() as colunas são separadas por tabulação
dados = pd.read_table('../01.Dados/dados.txt')
dados
```

```python
type(dados)
```

```python
from shapely.geometry import Point
```

```python
x = zip(dados.Longitude, dados.Latitude)
x
```

```python
list(x)
```

```python
geometry = [Point(x) for x in zip(dados.Longitude, dados.Latitude)]
```

```python
geometry
```

```python
crs = {'proj':'latlong', 'ellps':'WGS84', 'datum':'WGS84', 'no_defs':True}
geo_dados = gpd.GeoDataFrame(dados,crs=crs, geometry=geometry )
geo_dados
```

```python
type(geo_dados)
```

```python
import os

dir = '../01.Dados/Mapas/RJ-DATASET'
if not os.path.exists(dir):
    os.makedirs(dir)
```

```python
geo_dados.to_file(dir+'/DATASET.shp',encoding='UTF-8',driver='ESRI Shapefile')
```

```python
geo_dados.plot(figsize=(15,8),alpha=0.2)
```

**CRF** -> como é o sistema de coordenadas escolhido

**proj** - projeção do globo terreste, exmplo: 'latlong', 'utm', 'tmerc'

**ellps** - escolhe a forma da terra.

**datum** - onde determina eixo e origem do sistema de coordenadas específico. SIGAS2000, representa melhor a america latina
